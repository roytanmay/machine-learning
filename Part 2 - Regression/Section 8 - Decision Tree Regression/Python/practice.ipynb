```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('Position_Salaries.csv')

# Make sure that X is always represented as matrix and y is as vector
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X, y)
```

```python
y_pred = regressor.predict([[6.5]])
print(y_pred)
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, regressor.predict(X), color='green')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
# Decision Tree Regression model in non-continuous 
# In the above plot it is placing only the mentioned x values 
# We have to specity intervals for better representation

X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color='green')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
